# BTC OOS backtest report
Model trained with learning rate 0.01 and dropout 0.2 on the OOS split.

```python
import pandas as pd
from sklearn.metrics import accuracy_score, roc_auc_score
from backtest import load_returns, total_return, fmt_pct
```

```python
preds = pd.read_csv('runs/oos/preds.csv')
ret = load_returns('runs/oos/returns.csv')
```

```python
acc = accuracy_score(preds.y_true, preds.y_pred)
print('accuracy =', acc)
```

```python
auc = roc_auc_score(preds.y_true, preds.score)
print('auc =', auc)
```

```python
tot = total_return(ret)
print('total return =', fmt_pct(tot))
```
